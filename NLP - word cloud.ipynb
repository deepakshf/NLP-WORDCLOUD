# TEXT CLEANING

```python
text = " ".join(data).lower()
```

```python
def remove_emojis_and_symbols(text):
    emoji_pattern = re.compile("["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags (iOS)
        "\U00002702-\U000027B0"  # additional symbols
        "\U000024C2-\U0001F251"
        "\U0001F32D-\U0001F37F"                       
        "]+", flags=re.UNICODE)
    
    return emoji_pattern.sub(r'', text)
text=remove_emojis_and_symbols(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width = 1920, height = 1080, 
                      background_color ='white', 
                      stopwords = set(['the', 'and', 'to', 'a', 'of', 'in', 'is', 'for', 'that', 'I', 
    'with', 'this', 'it', 'are', 'on', 'as', 'have', 'was', 'they', 'my', 
    'at', 'has', 'be', 'from', 'which', 'an', 'by', 'their', 'you', 'all', 
    'but', 'not', 'been', 'can', 'we', 'me', 'so', 'also', 'university', 
    'faculty', 'students', 'experience', 'AAFT', 'media', 'industry', 
    'career', 'education', 'skills', 'opportunities']), 
                      min_font_size = 10).generate(text)

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0)
plt.show()
```
